```python
# Capture all columns failing 1NF (holds multiple values in a column)
cols_to_extract = ['PurchaseHistory', 'SubscriptionDetails',
       'ServiceInteractions', 'PaymentHistory', 'WebsiteUsage',
       'ClickstreamData', 'EngagementMetrics', 'Feedback',
       'MarketingCommunication']
```

```python
# preview an instance to understand steps needed for separation
customer_data['PurchaseHistory'].iloc[0]
```

```python
# library to convert from string to py object
import ast

ast.literal_eval(customer_data['PurchaseHistory'].iloc[0]) # works fine
```

```python
# convert string literal in customer_data df to py objects

def convert_string(col):
    if isinstance(col, str):
        return ast.literal_eval(col)
    else:
        return col
    
for col in cols_to_extract:
    customer_data[col].apply(convert_string)
```

```python
# verify

for col in cols_to_extract:
    print(f"{col}: \n {customer_data[col].iloc[0]}\n")
```

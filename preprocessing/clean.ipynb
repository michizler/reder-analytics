```python
# function to normalize nested columns
def normalize_column(df, column, id_col="CustomerID"):
    """
    Normalizes a column containing either:
    - dict  -> 1-to-1 table
    - list of dicts -> 1-to-many table

    ***args:
    - df -> dataframe with nested columns
    - column -> nested column to normalize
    - id_col -> column from dataframe to serve as primary key in resulting table
    """
    
    temp_df = df[[id_col, column]].dropna() # so there is no null value in a row

    first_value = temp_df[column].iloc[0]

    # For list of dictionaries
    if isinstance(first_value, list):
        temp_df = temp_df.explode(column)

        normalized = pd.json_normalize(temp_df[column])

        result = pd.concat(
            [temp_df[[id_col]].reset_index(drop=True),
             normalized.reset_index(drop=True)],
            axis=1
            )

    # For single dictionary
    elif isinstance(first_value, dict):
        normalized = pd.json_normalize(temp_df[column])
        
        result = pd.concat(
            [temp_df[[id_col]].reset_index(drop=True),
             normalized.reset_index(drop=True)],
            axis=1
            )

    else:
        raise TypeError(f"Unsupported data type in column '{column}': Must be a list or dict")

    # return and drop old index
    return result.reset_index(drop=True) # reset_index will replace current index with new default index: 0, 1, 2, ...
```

## Feature Engineering

### Purchase Behavior Features

**Idea**: Customers who spend more and buy consistently are less likely to churn.

- Total/average spending (loyalty indicator)
- Purchase consistency (stable vs erratic buyers)
- Product diversity (engaged vs single-product users)

```python
purchase_hist = purchase_hist.groupby('CustomerID').agg({
    'Value': 'sum',
    'Frequency': 'sum',
    'Product': 'nunique'
}).reset_index()
purchase_hist.columns = ['CustomerID', 'total_purchase_value', 'total_frequency', 'product_diversity']
```

```python
purchase_hist
```

### Subscription Lifecycle Features

**Idea**: Long-term subscribers with active accounts are less likely to churn.

- How long they've been subscribed (tenure)
- Whether subscription is still active
- Subscription plan type

```python
sub_details['Start_Date'] = pd.to_datetime(sub_details['Start_Date'])
sub_details['End_Date'] = pd.to_datetime(sub_details['End_Date'])
sub_details['subscription_duration_days'] = (sub_details['End_Date'] - sub_details['Start_Date']).dt.days
sub_details['subscription_age_days'] = (pd.to_datetime('2023-01-01') - sub_details['Start_Date']).dt.days
sub_details['is_active'] = (sub_details['End_Date'] > pd.to_datetime('2023-01-01')).astype(int)
sub_details = sub_details[['CustomerID', 'Plan', 'subscription_duration_days', 'subscription_age_days', 'is_active']]
```

```python
sub_details
```

### Service Interaction Patterns

**Idea**: Frequent support contacts (especially calls) may signal dissatisfaction.

- Total interactions
- Calculate last interaction date

```python
srv_interac['Date'] = pd.to_datetime(srv_interac['Date'])
srv_interac = srv_interac.groupby('CustomerID').agg({
    'Date': 'max',
    'Type': 'count'
}).reset_index()
srv_interac.columns = ['CustomerID', 'last_interaction_date', 'total_interactions']
```

```python
srv_interac
```

### Payment Reliability Features

**Idea**: Late payments indicate financial stress or dissatisfaction.

- Total late payments
- Late payment rate (percentage of payments that are late)
- Payment risk score (combined metric)

```python
payment_hist = payment_hist.groupby('CustomerID').agg({
    'Late_Payments': 'sum',
    'Method': 'count'
}).reset_index()
payment_hist.columns = ['CustomerID', 'total_late_payments', 'payment_count']
payment_hist['late_payment_rate'] = payment_hist['total_late_payments'] / (payment_hist['payment_count'] + 1)
payment_hist['payment_risk_score'] = payment_hist['total_late_payments'] * payment_hist['late_payment_rate']
```

```python
payment_hist
```

### Digital Engagement Features

**Idea**: Active website users are more engaged and less likely to leave.

- Page views and time spent
- Engagement ratio (time per page)
- Engagement intensity (overall activity level)

```python
web_usage = web_usage.groupby('CustomerID').agg({
    'PageViews': 'sum',
    'TimeSpent(minutes)': 'sum'
}).reset_index()
web_usage['engagement_ratio'] = web_usage['TimeSpent(minutes)'] / (web_usage['PageViews'] + 1)
web_usage['engagement_intensity'] = web_usage['PageViews'] * web_usage['TimeSpent(minutes)']
```

```python
web_usage
```

### Clickstream Behavior Features

**Idea**: Specific actions (adding to cart, searching) show purchase intent.

- Total actions taken
- Cart conversion rate (add-to-cart actions)
- Search intensity (exploration behavior)
- Page diversity (breadth of interest)

```python
clk_stream = clk_stream.groupby('CustomerID').agg({
    'Action': ['count', lambda x: (x == 'Add to Cart').sum(), lambda x: (x == 'Search').sum(), lambda x: (x == 'Click').sum()],
    'Page': 'nunique'
}).reset_index()
clk_stream.columns = ['CustomerID', 'total_actions', 'add_to_cart_count', 'search_count', 'click_count', 'unique_pages']
clk_stream['cart_conversion_rate'] = clk_stream['add_to_cart_count'] / (clk_stream['total_actions'] + 1)
clk_stream['search_intensity'] = clk_stream['search_count'] / (clk_stream['total_actions'] + 1)
clk_stream['page_diversity'] = clk_stream['unique_pages'] / (clk_stream['total_actions'] + 1)
```

```python
clk_stream
```

### Engagement Frequency Features

**Idea**: Daily users are more engaged than monthly users.

We convert frequency categories to numeric scores:

- Daily = 30 (highest engagement)
- Weekly = 4
- Monthly = 1 (lowest engagement)

```python
frequency_map = {'Daily': 30, 'Weekly': 4, 'Monthly': 1}
eng_metrics = eng_metrics.groupby('CustomerID').agg({
    'Logins': 'sum',
    'Frequency': 'first'
}).reset_index()
eng_metrics['frequency_score'] = eng_metrics['Frequency'].map(frequency_map)
eng_metrics['engagement_score'] = eng_metrics['Logins'] * eng_metrics['frequency_score']
eng_metrics = eng_metrics[['CustomerID', 'Logins', 'frequency_score', 'engagement_score']]
```

```python
eng_metrics
```

### Sentiment & Feedback Features

**Idea**: Negative feedback and low ratings predict churn.

- Average rating
- Comment length (longer = more invested)
- Sentiment flags (positive/negative)

```python
feedback = feedback.groupby('CustomerID').agg({
    'Rating': 'mean',
    'Comment': lambda x: x.str.len().mean()
}).reset_index()
feedback.columns = ['CustomerID', 'avg_rating', 'avg_comment_length']
feedback['sentiment_score'] = feedback['avg_rating'] / 5
feedback['is_negative'] = (feedback['avg_rating'] <= 2).astype(int)
feedback['is_positive'] = (feedback['avg_rating'] >= 4).astype(int)
```

```python
feedback
```

### Marketing Response Features

**Idea**: Customers who ignore emails are disengaging.

- Email open rate
- Click rate
- Click-through rate (clicks per open)
- Overall marketing engagement

```python
mkt_com = mkt_com.groupby('CustomerID').agg({
    'Email_Sent': 'count',
    'Email_Opened': 'count',
    'Email_Clicked': 'count'
}).reset_index()
mkt_com.columns = ['CustomerID', 'emails_sent', 'emails_opened', 'emails_clicked']
mkt_com['open_rate'] = mkt_com['emails_opened'] / (mkt_com['emails_sent'] + 1)
mkt_com['click_rate'] = mkt_com['emails_clicked'] / (mkt_com['emails_sent'] + 1)
mkt_com['click_through_rate'] = mkt_com['emails_clicked'] / (mkt_com['emails_opened'] + 1)
mkt_com['marketing_engagement'] = mkt_com['open_rate'] * mkt_com['click_rate']
```

```python
mkt_com
```

### Demographics & Temporal Features

**Idea**: Age, segment, and account age affect churn likelihood.

Would be nice to have:

- Age groups (Young, Adult, Middle, Senior)
- NPS categories (Detractor, Passive, Promoter)
- Account age in days

```python
customer_data['Timestamp'] = pd.to_datetime(customer_data['Timestamp'])
customer_data['account_age_days'] = (pd.to_datetime('2023-01-01') - customer_data['Timestamp']).dt.days
customer_data['age_group'] = pd.cut(customer_data['Age'], bins=[0, 25, 35, 50, 100], labels=['Young', 'Adult', 'Middle', 'Senior'])
customer_data['nps_category'] = pd.cut(customer_data['NPS'], bins=[-1, 6, 8, 10], labels=['Detractor', 'Passive', 'Promoter'])
demographic_features = customer_data[['CustomerID', 'Age', 'Gender', 'Segment', 'NPS', 'account_age_days', 'age_group', 'nps_category']]
```

```python
demographic_features
```

## Merging All Features

```python
customer_main = demographic_features.merge(purchase_hist, on='CustomerID', how='left') \
    .merge(sub_details, on='CustomerID', how='left') \
    .merge(srv_interac, on='CustomerID', how='left') \
    .merge(payment_hist, on='CustomerID', how='left') \
    .merge(web_usage, on='CustomerID', how='left') \
    .merge(clk_stream, on='CustomerID', how='left') \
    .merge(eng_metrics, on='CustomerID', how='left') \
    .merge(feedback, on='CustomerID', how='left') \
    .merge(mkt_com, on='CustomerID', how='left')

customer_main['ChurnLabel'] = customer_data['ChurnLabel']
```

### RFM Analysis

Segment customers based on rfm score. We'd determine rfm score by summing Recency, Frequency and Monetary.

```python
present_date = customer_main["last_interaction_date"].max()

customer_main['recency'] = present_date - customer_main['last_interaction_date']
customer_main['frequency'] = customer_main['total_frequency'] * customer_main['Logins'] # in case of both physical and platform interactions.
customer_main['monetary'] = customer_main['total_purchase_value']

R_label = range(5, 0, -1)
F_label = range(1, 6)
M_label = range(1, 6)

customer_main['recency'] = pd.qcut(customer_main['recency'], 5, labels=R_label)
customer_main['frequency'] = pd.qcut(customer_main['total_frequency'], 5, labels=F_label)
customer_main['monetary'] = pd.qcut(customer_main['total_purchase_value'], 5, labels=M_label)

customer_main[['recency', 'frequency', 'monetary']] = customer_main[['recency', 'frequency', 'monetary']].astype('int64')
customer_main['RFM_Score'] = customer_main['recency'] + customer_main['frequency'] + customer_main['monetary']

bins = [1, 4, 10, 15]
bin_labels = ['risk', 'engaged', 'loyalist']

customer_main['customer_segment'] = pd.cut(customer_main['RFM_Score'], bins=bins, labels=bin_labels, include_lowest=True)
customer_main.drop(columns=['RFM_Score', 'recency', 'frequency', 'monetary'], inplace=True)
```

```python
# Capture all columns failing 1NF (holds multiple values in a column)
cols_to_extract = ['PurchaseHistory', 'SubscriptionDetails',
       'ServiceInteractions', 'PaymentHistory', 'WebsiteUsage',
       'ClickstreamData', 'EngagementMetrics', 'Feedback',
       'MarketingCommunication']
```

```python
# library to convert from string to py object
import ast

ast.literal_eval(customer_data['PurchaseHistory'].iloc[0]) # works fine
```

```python
# convert string literal in customer_data df to py objects

def convert_string(col):
    if isinstance(col, str):
        return ast.literal_eval(col)
    else:
        return col
    
for col in cols_to_extract:
    customer_data[col] = customer_data[col].apply(convert_string)
```

```python
# function to normalize nested columns
def normalize_column(df, column, id_col="CustomerID"):
    """
    Normalizes a column containing either:
    - dict  -> 1-to-1 table
    - list of dicts -> 1-to-many table

    ***args:
    - df -> dataframe with nested columns
    - column -> nested column to normalize
    - id_col -> column from dataframe to serve as primary key in resulting table
    """
    
    temp_df = df[[id_col, column]].dropna() # so there is no null value in a row

    first_value = temp_df[column].iloc[0]

    # For list of dictionaries
    if isinstance(first_value, list):
        temp_df = temp_df.explode(column)

        normalized = pd.json_normalize(temp_df[column])

        result = pd.concat(
            [temp_df[[id_col]].reset_index(drop=True),
             normalized.reset_index(drop=True)],
            axis=1
            )

    # For single dictionary
    elif isinstance(first_value, dict):
        normalized = pd.json_normalize(temp_df[column])
        
        result = pd.concat(
            [temp_df[[id_col]].reset_index(drop=True),
             normalized.reset_index(drop=True)],
            axis=1
            )

    else:
        raise TypeError(f"Unsupported data type in column '{column}': Must be a list or dict")

    # return and drop old index
    return result.reset_index(drop=True) # reset_index will replace current index with new default index: 0, 1, 2, ...
```

```python
# create separate tables(dfs) for each nested column

sub_details = normalize_column(customer_data, "SubscriptionDetails")
srv_interac = normalize_column(customer_data, "ServiceInteractions")
payment_hist = normalize_column(customer_data, "PaymentHistory")
web_usage = normalize_column(customer_data, "WebsiteUsage")
clk_stream = normalize_column(customer_data, "ClickstreamData")
eng_metrics = normalize_column(customer_data, "EngagementMetrics")
feedback = normalize_column(customer_data, "Feedback")
mkt_com = normalize_column(customer_data, "MarketingCommunication")

arr = [sub_details, srv_interac, payment_hist, web_usage, clk_stream, eng_metrics, feedback, mkt_com]

for item in arr:
    print(f"{item}\n")
```

```python
customer_base = customer_data.drop(columns=cols_to_extract)

customer_base.head()
```

```python
# create one-to-one object that can be joined safely to customer dimension table
one_to_one = {
    "SubscriptionDetails": sub_details,
    "WebsiteUsage": web_usage,
    "EngagementMetrics": eng_metrics,
    "Feedback": feedback
}

# create new df for merge operation
customer_info = customer_base.copy()

# merge one-to-one dfs to customer_info
for df in one_to_one.values():
    customer_info = customer_info.merge(
        df,
        on="CustomerID",
        how="left"
    )
```
